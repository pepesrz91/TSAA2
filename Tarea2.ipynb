### 2) Aplica los algoritmos de stemming vistos en clase al primer comentario que se tiene en review01 (es decir, solo utiliza el comentario que inicia “These flannel wipes are OK...”) y elije el que consideres el mejor resultado por simple inspección visual.



```
import nltk
nltk.download()
```

```
from nltk.stem import PorterStemmer
from nltk.stem import LancasterStemmer
```

```
porter = PorterStemmer()
lancaster=LancasterStemmer()
```

```

porter_stem_words = []
for w in review_01[0]:
  temp = porter.stem(w)
  porter_stem_words.append(temp)
```

```
print(porter_stem_words)
```

```

lancaster_stem_words = []
for w in review_01[0]:
  temp = lancaster.stem(w)
  lancaster_stem_words.append(temp)
```

```
print(lancaster_stem_words)
```

### El método de lancaster da resultados más concretos y logra reducir mucho más el tamaño de la información.
